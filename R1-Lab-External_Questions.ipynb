### Multiply Sepal Length and width and store it under the column name "SepalExtra" in the same Iris dataframe

```python
iris = pd.read_csv("Iris.csv")

iris.rename(columns={'Sepal Length (in cm)':'sepalL',
                     'Sepal Width in (cm)':'sepalW',
                     'Petal length (in cm)':'petalL',
                     'Petal width (in cm)':'PetalW'}, 
                 inplace=True)

sepalE = iris.sepalL * iris.sepalW

iris['SepalExtra'] = sepalE

iris.head()
```
